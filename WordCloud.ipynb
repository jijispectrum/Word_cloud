```python
#Worcloud
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
df = pd.read_csv("android-games.csv")
df
```

```python
df.head()
#Checking for NaN values
df.isna().sum()
```

```python
# Concatenate text data from the column of interest (e.g., 'title' or 'category')
text_data = ' '.join(df['category'])
text_data
```

```python
# Generate the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text_data)
# Change the colormap of the word cloud to 'twilight'
wordcloud.colormap = 'PiYG'
```

```python
# Plot the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
